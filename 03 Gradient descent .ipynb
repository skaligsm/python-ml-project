# Example 1 --A simple cost function

$$f(x) = x^2+x+1$$

```python
def f(x):
    return x**2 +x + 2
```

```python
#Make Data
x_1 = np.linspace(start=-3,stop=3,num=500)
```

```python
#Plot
plt.xlim([-3,3])
plt.ylim(0,8)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)
plt.plot(x_1,f(x_1))
plt.show()
```

## Slope & Derivatives

```python
def df(x):
    return 2*x+1
```

```python
#Plot function and derivative side by side
plt.figure(figsize=[15,5])

# 1 Chart:Cost function
plt.subplot(1,2,1)

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function',fontsize=17)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

plt.plot(x_1,f(x_1),color="blue",linewidth=3)

#2 Chart: Derivative
plt.subplot(1,2,2 )

plt.title('Slop of the cost function',fontsize=17)

plt.xlabel('x',fontsize=16)
plt.ylabel('df(x)',fontsize=16)

plt.grid()

plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1 ),color='skyblue',linewidth=5)
plt.show()
```

 ## Python Loops & Gradient Descent

```python
#Python For Loop
for n in range(5):
    print('Hey ..',n)
    
print('End of loop')    
```

```python
#Pythin while loop
n = 0 
while n < 5:
    print('Count..', n)
    n= n + 1
print('End ..')   
```

```python
#Gradient Dascent
new_x = 3
previous_x = 0
step_multiplier = 0.1
precision = 0.0001

x_list = [new_x]
slop_list = [df(new_x)]

for n in range(5000):
    previous_x = new_x
    gradient = df(previous_x)
    new_x = previous_x - step_multiplier *gradient
    
    step_size = abs(new_x-previous_x)
    #print(step_size)
    
    x_list.append(new_x)
    slop_list.append(df(new_x))
    
    if step_size < precision:
        print('Loop run this many times:',n)
        break
    
print('Local minimum occurs at:',new_x)  
print('Slop or df(x) value at this point is',df(new_x))
print('f(x) value or cost at this point is:',f(new_x))
    
```

```python
# Super impose the gradient dascent calculations on plot
plt.figure(figsize=[15,5])

# 1 Chart:Cost function
plt.subplot(1,2,1)

plt.xlim([-3,3])
plt.ylim(0,8)

plt.title('Cost function',fontsize=17)
plt.xlabel('x',fontsize=16)
plt.ylabel('f(x)',fontsize=16)

plt.plot(x_1,f(x_1),color="blue",linewidth=3,alpha=.8)

values = np.array(x_list)
plt.scatter(x_list,f(values), color='red',s=100,alpha=.6)

#2 Chart: Derivative
plt.subplot(1,2,2 )

plt.title('Slop of the cost function',fontsize=17)

plt.xlabel('x',fontsize=16)
plt.ylabel('df(x)',fontsize=16)

plt.grid()

plt.xlim(-2,3)
plt.ylim(-3,6)

plt.plot(x_1,df(x_1 ),color='skyblue',linewidth=5,alpha= .6)
plt.scatter(x_list,slop_list,color='red',s=100,alpha=.5)
plt.show()
```

## Gradient Dascent as a Python Function

```python
def gradient_dascent(derivative_func, initial_guess, multiplier = 0.02 ,precision = 0.001 ):

                new_x = initial_guess
                x_list = [new_x]
                slop_list = [derivative_func(new_x)]

                for n in range(5000):
                    previous_x = new_x
                    gradient = derivative_func(previous_x)
                    new_x = previous_x - multiplier *gradient

                    step_size = abs(new_x - previous_x)
                    #print(step_size)

                    x_list.append(new_x)
                    slop_list.append(derivative_func(new_x))

                    if step_size < precision:
                        break
                
        
                return new_x, x_list, slop_list      
```

```python
local_min, list_x, deriv_list = gradient_dascent(dg,0.5, 0.02,0.001)
print('Local min occurs at:',local_min)
print('Number of steps:',len(list_x))
```

```python
local_min, list_x, deriv_list = gradient_dascent(derivative_func= dg,initial_guess= - 0.1)
print('Local min occurs at:',local_min)
print('Number of steps:',len(list_x))
```
